# Задание №3

## Что такое Немезида, и как она работает?
Система, которая борется с зарабатыванием на копировании чужого контента во ВКонтакте. Работает на основе нейронной сети. 
Ее рабочий процесс выглядит так:

Сеть проверяет любой контент в ВК на оригинальность т.е. ищет похожий контент в ВК и в случае подозрения на неоригинальность контента отправляет предупреждения модераторам -- > модераторы рассматривают предупреждение от системы -- > при необходимости модераторы делают предупреждение тому, кто опубликовал потенциально неоригинальный контент.

## Какие риски и негативные последствия для нас?
ВКонтакте обещает максимально осложнить заработок для страниц, которые зарабатывают на копировании чужого контента. При получении 5 предупрждений за неделю, обещают штрафовать понижением в рейтинге рекомендаций. При получении 15 предупреждений за неделю обещают даже отключить от систем монетизации в ВК.

## Как нейтрализовать риски и негативные последствия?
К счастью, мы можем нейтрализовать риски и негавтиные последствия. Для этого следует учитывать рекомендации, которые дает сам ВКонтакте:
- Продолжать создавать уникальный контент, который создаем мы и наши подписчики.
- Размещать контент в ВК максимально быстро, раньше, чем на других платформах. Если кто-то разместит в ВК НАШ контент из любого другого источника, кроме ВК, то этот кто-то будет первоисточником для Немезиды, а не мы. ВКонтакте обещает, что наказывать сообщество за это не будет, но это потенциальная потеря времени и денег при разбирательствах.
- При репосте других записей следует добавлять комментарии к репосту. Это особенно важный момент для нас, потому что мы делаем репосты материалов из наших тематических страниц в ВК на главной странице, и наооборот.
- Указывать ссылки на настоящих авторов. Уважаем авторское право.
- Оперативно реагировать на предупреждения от модераторов ВК, разбираться с ними.

## Какие преимущества и положительные последствия для нас?
Если система действительно начнет работать правильно и поможет убрать плагиат из ВК, то мы будем:
- Более уверены в том, что наш контент не буду воровать.
- Легче пробираться в топ по рекомендациям для потенциальных подписчиков, потому что мы будем выделяться как оригинальный источник контента. А уже чем больше подписчиков, тем больше охват и больше заработанных денег.
